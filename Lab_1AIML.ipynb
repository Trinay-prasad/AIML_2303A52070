```python
people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 100]
weight=[45, 55, 65, 75]
import matplotlib.pyplot as plt
plt.scatter(weight, height)
plt.title("Relationship between weight and height")
plt.ylabel("Height in Meters")
plt.xlabel("weight of patients in kg")
plt.show()

#bar chart
plt.bar(people,weight)
plt.xlabel("people")
plt.ylabel("weight")
plt.title("Bar Chart of weight for each Person")
plt.show()

```

```python
plt.hist(age,bins=5)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("Histogram of age")
plt.show()
```
